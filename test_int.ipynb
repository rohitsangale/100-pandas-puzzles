### Calculate the portfolios excess returns (portfolio return – benchmark return)

```python
data['P1_excess']=data['P1']-data['MSCI World']
data['P2_excess']=data['P2']-data['MSCI World']
data['P3_excess']=data['P3']-data['MSCI World']
data['P4_excess']=data['P4']-data['MSCI World']
data['P5_excess']=data['P5']-data['MSCI World']

data
```

### Calculate the yearly excess return per portfolio (e.g. return for 2010, 2011, etc.)

```python
data['Year']=data.index.year
```

```python
def cumulative_excess_return(monthly_returns):
    return (np.prod(1 + monthly_returns) - 1)
```

```python
data.groupby('Year')['P1_excess'].apply(cumulative_excess_return).reset_index()
```

```python
data.groupby('Year')['P2_excess'].apply(cumulative_excess_return).reset_index()
```

```python
excess = [col for col in data.columns if col.endswith('excess')]
excess
```

```python
combined=pd.DataFrame({'Year':range(2010,2019)})
for i in excess:
  x=data.groupby('Year')[i].apply(cumulative_excess_return).reset_index()
  combined=pd.merge(combined,x,on='Year',how='outer')
combined


```
